# Another Example

```python
data = {'customer':['Jones Brothers','Beta Corp','Globex Corp','Acme','Initech','Hooli'],
        'sales'   :[500,'$1,000.00',300.1,'$750.01',300,250]}

```

```python
df = pd.DataFrame.from_dict(data,orient='columns')
```

```python
df
```

we have unkonow characters like "$ and , 

```python
df['Sales_Type'] = df['sales'].apply(lambda x: type(x).__name__)
```

```python
df
```

```python
def clean_currency(x):
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)
```

If the value is a string, then remove currency symbol and delimiters otherwise, the value is numeric and can be converted

```python
df['sales'] = df['sales'].apply(clean_currency).astype('float')
df['Sales_Type'] = df['sales'].apply(lambda x: type(x).__name__)
```

```python
df
```

```python
df.drop('Sales_Type',axis=1,inplace=True)
```
